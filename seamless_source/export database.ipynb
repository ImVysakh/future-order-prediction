```python
import mysql.connector as connection
import pandas as pd
```

```python
sql = 'select * from demoss.articles;'
```

```python
db = connection.connect(host='localhost',database='demoss',user='root',password='root',use_pure=True)
```

```python
df = pd.read_sql(sql,db)
```

```python
def execute(c, command):
    c.execute(command)
    return c.fetchall()
```

```python
c = db.cursor()
```

```python
for table in execute(c, "show tables;"):
    table = table[0]
    cols = []
    for item in execute(c, "show columns from " + table + ";"):
        cols.append(item[0])
    data = execute(c, "select * from " + table + ";")
    with open(table + ".csv", "w", encoding="utf-8") as out:
        out.write("\t".join(cols) + "\n")
        for row in data:
            out.write("\t".join(str(el) for el in row) + "\n")
    print(table + ".csv written")
```

```python
import csv

for table in execute(c, "show tables;"):
    table = table[0]
    cols = []
    for item in execute(c, "show columns from " + table + ";"):
        cols.append(item[0])
    data = execute(c, "select * from " + table + ";")
    with open(table + ".csv", "w", encoding="utf-8", newline='') as out:
        writer = csv.writer(out, delimiter=',')
        writer.writerow(cols)
        for row in data:
            writer.writerow(row)
    print(table + ".csv written")
```
